```python
import numpy as np
import pandas as pd 
import sklearn
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
left_category = set([i for i in range(1,26)])
def compute_y(row, alpha):
    x1, x2 = row['x1'], row['x2']
    noise = np.random.normal()
    if x1>0 and x2 in left_category:
        return alpha + noise
    if x1 <= 0 and x2 not in left_category:
        return 1 + noise
    return noise

df = pd.DataFrame()
df['x1'] = np.random.normal(0,1,300)
df['x2'] = np.random.randint(1,50,300)
```

```python
from sklearn.tree import DecisionTreeRegressor
# for alpha in [2]:
alpha = 2
df['y'] = df.apply(lambda row : compute_y(row,alpha), axis = 1)
regressor = DecisionTreeRegressor(random_state=0,max_depth = 4)
n_splits = 5
    
    
```
