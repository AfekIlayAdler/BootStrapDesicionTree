```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from glob import glob
from pathlib import Path
```

```python
kfold = glob("k*60.csv")
reg = glob("r*60.csv")

kfold_mean = pd.concat([pd.read_csv(j,index_col = 0) for j in kfold], axis = 1).mean(axis = 1)
normalized_series = pd.Series(kfold_mean/kfold_mean.sum())
print(normalized_series)
normalized_series.plot(kind = 'bar')
plt.show()

reg_mean = pd.concat([pd.read_csv(j,index_col = 0) for j in reg], axis = 1).mean(axis = 1)
normalized_series = pd.Series(reg_mean/reg_mean.sum())
print(normalized_series)
normalized_series.plot(kind = 'bar')
plt.show()
```
